```python
# imports
import os 
from dotenv import load_dotenv
import anthropic
from scraper import fetch_website_contents
```

```python
# set up environment
load_dotenv(override=True)
api_key = os.getenv('ANTHROPIC_API_KEY')
client = anthropic.Anthropic()
```

```python
def ask_claude(question):
    messages = [{"role": "user", "content": question}]
    response = client.messages.create(
        model="claude-haiku-4-5",
        max_tokens=1000,
        messages=messages,
    )
    return response.content[0].text
```

```python
ask_claude("What is the capital of Italy?")
```

```python
system_prompt = """
You are a professional data analyst, with years of expertise and experience
in creating short summaries of length articles, blogs etc.
"""
```

```python
system_prompt = "You are a professional summarizer. Summarize the user's text in exactly two sentences. Respond in plain prose."

def summarize(text):
    messages = [{"role": "user", "content": text}]
    response = client.messages.create(
        model="claude-haiku-4-5",
        max_tokens=500,
        system=system_prompt,
        messages=messages,
    )
    return response.content[0].text
    

print(summarize("Optical Character Recognition (OCR) is a technology that converts images of text — scanned invoices, photos of documents, PDFs, screenshots, handwritten forms — into machine-readable text."))
```
